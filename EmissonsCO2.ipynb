```python
df2.replace('...', np.nan, inplace=True)

# Listar as colunas do DataFrame antes da conversão
print("Tipos de dados antes da conversão:")
print(df2.dtypes)

# Converter todas as colunas para float, exceto 'Country_nan'
colunas_float = df2.columns.difference(['Country_nan'])
df2[colunas_float] = df2[colunas_float].astype(float)

# Listar as colunas do DataFrame depois da conversão
print("\nTipos de dados depois da conversão:")
print(df2.dtypes)
```

```python
# Identificar valores que não podem ser convertidos para float
valores_nao_numericos = df2.loc[~pd.to_numeric(df2['CO2 emissions _mio. tonnes '], errors='coerce').notna(), 'CO2 emissions _mio. tonnes ']

# Exibir valores não numéricos
print("Valores não numéricos na coluna {}: \n{}".format('CO2 emissions _mio. tonnes ', valores_nao_numericos))

# Corrigir manualmente os valores não numéricos e converter a coluna para float
df2['CO2 emissions _mio. tonnes '] = pd.to_numeric(df2['CO2 emissions _mio. tonnes '], errors='coerce')

# Exibir o DataFrame depois da verificação
print("\nTipos de dados depois da verificação:")
print(df2.dtypes)
print("\nDataFrame depois da verificação:")
print(df2)
```

**FAZENDO O MAPA COM AS INFORMAÇÕES DE EMISSÕES**

```python
import plotly.graph_objs as go
import plotly.offline as offline

# Lista para armazenar os dados do slider
data_slider = []

# Itera sobre os dados e adiciona ao slider
for i in range(len(df2)):
    trace = go.Choropleth(
        z=[df2['CO2 emissions per capita_tonnes'][i]],
        locations=[df2['Country_nan'][i]],
        text=[df2['Country_nan'][i]],
        zmin=0,
        zmax=df2['CO2 emissions per capita_tonnes'].max(),
        colorscale='Viridis',
        colorbar=dict(title='CO2 Emissions (tons per capita)')
    )
    data_slider.append(trace)

# Configuração do slider
steps = []
for i in range(len(df2)):
    step = dict(
        method='restyle',
        args=['visible', [False] * len(df2)],
        label=str(df2['Country_nan'][i])
    )
    step['args'][1][i] = True
    steps.append(step)

sliders = [dict(active=0, pad={"t": 1}, steps=steps)]

# Layout do gráfico
layout = dict(
    geo=dict(
        scope='world',
        projection={'type': 'mercator'},
        lataxis_range=[-32, 80],
        lonaxis_range=[-168, 184],
        showlakes=False,
        showrivers=False
    ),
    title=dict(
        text='CO2 emissions (tons per capita)',
        font=dict(family="Times New Roman", size=40),
        x=0.02,
        y=0.90
    ),
    autosize=False,
    margin=dict(l=20, r=10, t=50, b=20),
    paper_bgcolor="white",
    width=800,
    height=600,
    sliders=sliders,
)

# Criação do gráfico
fig = dict(data=data_slider, layout=layout)

# Exibição do gráfico
offline.iplot(fig, config={'scrollZoom': False, 'modeBarButtonsToRemove': ['pan2d', 'zoomIn2d', 'zoomOut2d', 'autoScale2d', 'resetScale2d']})
```
